```python
def fill_null_data(data):

    categorical_cols , numerical_cols = [], []
    # checking the type of the data and appending them into respective lists.
    try:
        for col in data.columns:
            if data[col].dtypes == "object":
                print("The {} is a categorical_data".format(col))
                categorical_cols.append(col)
            elif data[col].dtypes != "object":
                print("The {} is a numerical_data".format(col))
                numerical_cols.append(col)
        print("categorical cols are:", categorical_cols)
        print("numerical cols are:", numerical_cols)
        for col in categorical_cols:
            if data[col].isna().sum() == 0:
                continue
            elif data[col].isna().sum() != 0:
                print("The col {} has null data".format(col))
                data[col].fillna(data[col].mode()[0], inplace = True)
                print("The null data has been successfully imputed for the column {}".format(col))

        for col in numerical_cols:
            if data[col].isna().sum() == 0:
                continue
            elif data[col].isna().sum() != 0:
                print("The col {} has null data".format(col))
                data[col].fillna(data[col].mean(), inplace = True)
                print("The null data has been successfully imputed for the column {}".format(col))
    except AssertionError as msg:
        print(msg)
    
    return categorical_cols , numerical_cols
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for cols in categorical_cols:
    df[cols] = encoder.fit_transform(df[cols])
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
# Separating the features (X) and the target (y)
X = df.drop(["SALES"], axis =1)
y = df["SALES"]

# Defining the training and the test set
X_train , X_test , y_train , y_test = train_test_split(X,y, train_size = 0.8)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Scaling the features
scaler.fit_transform(X_train)
scaler.transform(X_test)
```

### **Calling Linear Regresson Model**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Start the training
model.fit(X_train, y_train)
```

```python
# Making Predictions:
y_pred = model.predict(X_test)
```
