```python
numerical_cols, categorical_cols = [],[]
for column in data.columns:
    print(column)
    print({column:data[column].dtypes})
    if data[column].dtypes == "object":
        categorical_cols.append(column)
    else:
        numerical_cols.append(column)
```

```python
# Automated plots for categorical data / Univariate
for feature in numerical_cols:
    plt.figure(figsize  = [10,8])
    ax = sns.histplot(data[feature], color = "darkblue")
    for bars in ax.containers:
      ax.bar_label(bars)
    plt.title("Freq. Distribution for {}".format(feature))
    plt.xlabel(feature)
    plt.ylabel("Freq. of {}".format(feature))
    plt.grid()
```

## **Label Encoding**

```python
# Creating another data copy
data_3 = data.copy()
```

```python
data_3.head()
```

```python
data_3.isnull().sum()
```

```python
# Here sex column can be labelled or converted to numerical column.
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
# Labelling Sex column
data_3["Sex"] = encoder.fit_transform(data_3["Sex"])
# So male is labelled as 1 and female is labelled as 0.
```

```python
data_3.head()
```

```python
# Same we can perform label encoding for cabin column.
data_3["Cabin"] = encoder.fit_transform(data_3["Cabin"])
```

```python
data_3.head()
```
