# Filtering and Ordering Data

```python
df = pd.read_csv("world_population.csv")
```

```python
specific_countries = ['India', 'Russia']

df[df["Country"].isin(specific_countries)]
```

## Pandas Group by

```python
flavours = pd.read_csv("Flavors.csv")
```

```python
flavours.head(10)
```

```python
flavours.groupby('Base Flavor').mean()
```

```python
flavours.groupby('Base Flavor').count()
```

```python
flavours.groupby('Base Flavor').min()
```

```python
flavours.groupby('Base Flavor').max()
```

```python
flavours.groupby('Base Flavor').agg({'Texture Rating':['mean', 'max','min','sum'], 'Flavor Rating': ['mean', 'max','min','sum']})
```

```python
flavours.groupby(['Base Flavor','Liked']).mean()
```
