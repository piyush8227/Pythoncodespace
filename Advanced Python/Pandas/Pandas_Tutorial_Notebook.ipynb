#### **Data Slicing with dataframe.loc() and dataframe.iloc()**

- **`loc`** is used to index a pandas DataFrame or Series using **labels**.

- On the other hand, **`iloc`** can be used to retrieve records based on their positional index.

- However, both are used for **slicng w.r.t index.**

<center><img src="https://miro.medium.com/max/1400/1*CgAWzayEQY8PQuMpRkSGfQ.png" width="900" height=450" /></center>

**Note:** 

- if we return **`data.loc[0]`** and **`data.iloc[0]`** we will get the same output.

- One core difference in between **`data.loc()`** and **`data.iloc()`** is that while slicing **`data.iloc()`** excludes the last element.

- However, **`data.loc()`** includes the last element.

- i.e, **`data.loc()`** does not follow the **principle of list indexing** whereas **`data.loc()`** follows the **principle of list indexing**.

- **`data.iloc()`** allows **only integers** to be passed within however, for **`data.loc()`** one can pass **integers** and **labels/columns names**.

```python
data.head(1)
```

```python
data.loc[0]
```

```python
data.iloc[0]
```

```python
data.columns
```

```python
# If I want all the rows for a specific label
data.loc[:,['Sales', 'Order Date']]
```

```python
data.iloc[:,[0, 1]]
```

```python
data.loc[[0,1,2],:]
```

**Note:**
- We can re-index the **`row sequence`** as well.

```python
data.loc[[0,2,1],:]
```

**Note:**
- For a **better practice** always try to specify the range.

```python
# Return all the columns for row 0,1 and 2
data.loc[0:2,:]
```

**Note:**
- If we are not specifying what columns we want pandas automatically considers **all the columns.**

- Similarly if we are not specifying what rows we want pandas automatically considers **all the rows.**

```python
data.loc[:2]
```

```python
# If I want all the rows for Sales and its corresponding Order ID
data.loc[:,['Order ID','Sales']]
```

**Note:**

- In the example below, notice how the slicing is computed.

- **`2:4`** does not correspond to indices but instead, to **`labels`**.

- In other words, it tells pandas to return **all the rows in between the indices 2 and 4**.

```python
data.loc[2:4]
```

**Note:**

- In the example below, notice how the **slicing is computed**.

- We got an **error.** 

- We tried to slice out the **1st and 17th column** from the dataframe **data** along with **all the corresponding rows** using the **iloc method.**

- However, as loc is **label name sensetive** i.e, **`data.loc() performs search operation by the name of the columns`**.

- Hence, we will have to pass the **label/column names** instead **of the index of the labels**.

```python
data.loc[:,[1,17]]
```

**Note:**

- The above issue can be resolved with data.iloc() method.

- **`data.iloc[:,[1,17]]`** will return the same results as **`data.loc[:,['Order ID','Sales']]`**.

- This is because data.iloc() is not **`label name sensetive`** instead it is **`label index sensetive`**.

- **`data.iloc() performs its search operation by the index of the columns.`**

- However, if we are passing non integers in the space of columns for e.g, data.iloc[:,['Order ID','Sales']] this throws an error as data.iloc searches by the index of columns.

```python
data.iloc[:,[1]]
```

```python
data.iloc[:,['Order ID','Sales']]
```

### **Pandas Groupby**

- **Pandas GroupBy** is a powerful and versatile function in Python.

- It allows you to split your data into separate groups to perform computations for better analysis.

<center><img src="https://cdn.analyticsvidhya.com/wp-content/uploads/2020/03/groupby_example.png" width="400" height=400" /></center>

- Let me take an example to elaborate on this.

-  Letâ€™s say we are trying to analyze the weight of a person in a city.

- We can easily get a fair idea of their weight by determining the **mean weight** of all the city persons.

- But hereâ€˜s a question â€“ **Would the weight be affected by the gender of a person?**

- We can group the city persons into different **gender groups** and calculate their **mean weight.**

- This would give us a better insight into the weight of a person living in the city. 

- But we can probably get an even better picture if we further separate these gender groups into different age groups and then take their mean weight (because a teenage boyâ€™s weight could differ from that of an adult male)!

```python
categorical_columns
```

```python
group_1 = data.groupby(by='State')
```

```python
group_1
```

**Observations:**

- **GroupBy** has conveniently returned a **DataFrameGroupBy** object.

- It has split the data into **separate groups**.

- **However, it wonâ€™t do anything unless it is being told explicitly to do so.**

```python
data.groupby(by='State').count()
```

```python
data[data['State']=='Arizona']
```
