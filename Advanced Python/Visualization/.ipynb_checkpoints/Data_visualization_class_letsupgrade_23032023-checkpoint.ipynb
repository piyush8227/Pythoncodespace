```python
numerical_cols, categorical_cols = [],[]
for column in data.columns:
    print(column)
    print({column:data[column].dtypes})
    if data[column].dtypes == "object":
        categorical_cols.append(column)
    else:
        numerical_cols.append(column)
```

```python
categorical_cols
```

```python
numerical_cols
```

```python
# Automated plots for categorical data / Univariate
for feature in categorical_cols:
    plt.figure(figsize  = [10,8])
    sns.countplot(data = data,
                  y = feature)
    plt.title("Countplot for {}".format(feature))
```

```python
# Automated plots for categorical data / Univariate
for feature in numerical_cols:
    plt.figure(figsize  = [10,8])
    plt.hist(data[feature], color = "darkblue")
    plt.title("Freq. Distribution for {}".format(feature))
    plt.xlabel(feature)
    plt.ylabel("Freq. of {}".format(feature))
    plt.grid()
```
