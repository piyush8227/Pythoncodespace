```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('/content/train_and_test2.csv')
```

```python
data['Pclass'].value_counts()
```

```python
data.groupby(['Pclass'])['2urvived'].count()
```

```python
sizes = data.groupby(['Pclass'])['2urvived'].count().values
labels = data.groupby(['Pclass'])['2urvived'].count().index
list1 = [0.1, 0, 0]
plt.figure(figsize = [16,8])
plt.pie(sizes,
        labels=labels,
        autopct='%1.0f%%',
        explode = list1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize = [16,8])
plt.grid()
sns.countplot(y = 'Pclass', hue = 'Sex', data = data, orient = 'h')
```

### **Box Plot and Violin Plots**

```python
data.columns
```

```python
sns.violinplot(x=data['Sex'])
plt.show()
```

```python
# y = 'Pclass', hue = 'Sex',
plt.figure(figsize= [16,8])
sns.violinplot(x='Sex',y = 'Pclass',data = data)
plt.show()
```

```python
# y = 'Pclass', hue = 'Sex',
plt.figure(figsize= [16,8])
sns.violinplot(x='Sex',y = 'Pclass',hue ='2urvived' ,data = data)
plt.show()
```

```python
# sns.violinplot(x='Sex',y = 'Pclass',hue ='2urvived' ,data = data)
plt.figure(figsize= [16,8])
sns.boxplot(x='Pclass',y ='2urvived',hue ='Sex',data = data, orient = 'h')
plt.show()
```
